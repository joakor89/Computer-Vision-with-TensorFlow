## Let's distribute in Training and Test Database 


*   Database distribute on Training & Test
*   Let's define the path where Training & Test Database will be stored.

```python
# General Folder Path
path_general = "final_database"
# Train Address
path_train = "{}/train".format(path_general)
#Test Address
path_test = "{}/test".format(path_general)

# Gegenerating Folder on Terminal Command
os.mkdir(path_general)
# Train Folder Generation
os.mkdir(path_train)
# Test Folder Generation
os.mkdir(path_test)
```

El factor es de 70% train y 30% test de manera aleatoria.

```python
# Random Library
import random
# Shutil Library allows to move object among folders
import shutil

# Let's initially store training data at 70 % 
train = 0.7
# Let's perambulate on content variable at 70 %
for nCount in range(int(len(content)*train)):
  # Let's random choice on list
  random_choice_img = random.choice(content)
  # Absolute Image Generation
  random_choice_img_abs = "{}/{}".format(path,random_choice_img)
  # Allocation folder
  target_img = "{}/{}".format(path_train, random_choice_img)
  # Copying files
  shutil.copyfile(random_choice_img_abs, target_img)
  # Removing arguments 
  content.remove(random_choice_img)
```

```python
# Storing the remaining data list as test data
for img in content:
  # Ramaining source's images
  random_choice_img_abs = "{}/{}".format(path, img)
  # Path test
  target_img = "{}/{}".format(path_test, img)
  # Copying files
  shutil.copyfile(random_choice_img_abs, target_img)
```
